```python

import time
from phue import Bridge

bridge = Bridge("192.168.178.21")

def light_dim(bridge, dim_value: int, step:int):
    print("Dimming lights to", dim_value)
    for light in bridge.lights:
        for bri in range(0, dim_value, step):
            print(bri)
            light.brightness = bri
            time.sleep(0.05)

def dim_room(bridge, room_name: str, dim_value: int, step:int):
    print("Dimming lights in room:", room_name, "to", dim_value)

    group_id = bridge.get_group_id_by_name(room_name)
    start_bri = 0 if step > 0 else dim_value
    end_bri   = dim_value if step > 0 else 0

    for bri in range(start_bri, end_bri, step):
        print(f"  → brightness {bri}")
        bridge.set_group(group_id, "bri", bri)
        time.sleep(0.05)
```

```python
import time
from phue import Bridge

b = Bridge("192.168.178.21")
b.connect()

light_id = 2            # change to your light id or use 'all'
step     = 10           # brightness decrement per step
pause    = 0.2          # seconds between steps

# start at full brightness
b.set_light(light_id, "bri", 254)

for bri in range(254, 0, -step):
    b.set_light(light_id, "bri", bri)
    time.sleep(pause)

# Optional: turn it off completely
b.set_light(light_id, "on", False)
```
