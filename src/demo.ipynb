```python
import time
from phue import Bridge

b = Bridge("192.168.178.21")
b.connect()

light_id = 2            # change to your light id or use 'all'
step     = 10           # brightness decrement per step
pause    = 0.2          # seconds between steps

# start at full brightness
b.set_light(light_id, "bri", 254)

for bri in range(254, 0, -step):
    b.set_light(light_id, "bri", bri)
    time.sleep(pause)

# Optional: turn it off completely
b.set_light(light_id, "on", False)
```
