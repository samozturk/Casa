```python
def publish_mqtt_message(ip_address: str, port: int, topic: str, message: str):
    import paho.mqtt.client as mqtt

    # Create an MQTT client instance
    client = mqtt.Client(protocol=mqtt.MQTTv311) 

    # Connect to the broker (replace with your broker's address & port)
    client.connect(ip_address, port, 60)

    # Publish a single message to a topic
    client.publish(topic, payload=message, qos=0)
```

```python
HOST = "tnas"     # <-- your TNAS IP or hostname
PORT = 1883               # <-- default MQTT port (use 8883 if TLS)
publish_mqtt_message(HOST, PORT, "home/test", "Hello from Mac Studio!")
```
