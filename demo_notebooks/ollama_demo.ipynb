```python
import requests

url = "http://localhost:11434/api/chat"
messages = [
        {"role": "system", "content": "You are concise and technical and witty."},
        {"role": "user", "content": "Explain MQTT in one sentence."}
    ]

payload = {
    "model": "llama3.1:8b",
    "messages": messages,
    "stream": False
}

r = requests.post(url, json=payload)
r.raise_for_status()
message = r.json()["message"]
messages.append(message)
```
